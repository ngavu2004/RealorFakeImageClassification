```python
import gradio as gr
import numpy as np

def predict(img):
  X = []
  X.append(cv2.resize(cv2.imread(img.name),dsize=(256,256)))
  X = np.array(X)
  X = X/255
  Resnet_pred = Resnet_based_model.predict(X)
  EfficientNet_pred = EfficientNet_based_model(X)
  Resnet_clipped_pred = np.clip(Resnet_pred,0.04,0.96)
  EfficientNet_clipped_pred = np.clip(EfficientNet_pred,0.04,0.96)
  ensemble_pred = 0.5*Resnet_clipped_pred + 0.5*EfficientNet_clipped_pred
  print(ensemble_pred[0][0])
  if float(ensemble_pred[0][0]) >= 0.5:
    return 'Fake Image: {}%'.format(ensemble_pred[0][0]*100)
  else:
    return 'Real image: {}%'.format((1-ensemble_pred[0][0])*100)

im = gr.inputs.Image(tool = None, image_mode='RGB', type='file')
iface = gr.Interface(
    fn = predict, 
    inputs = im, 
    outputs="text",
    interpretation="default"
)
iface.launch(debug=True)
```
